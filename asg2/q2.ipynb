Imports needed for different functionalities

```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
```

Plot of lambda vs g 

```python
# Defining the activtion function that is to be used
def act(x):
    return np.tanh(x)
```

```python
#Defining the values of lambda and g for the plot The 
lam = np.linspace(-1,1,1000)
g = np.linspace(-1,1,1000)
L,G = np.meshgrid(lam,g)
f = act(L*G)
# Defineing how to plot the function
fig1 = plt.figure()
ax = fig1.gca(projection='3d')
ax.plot_surface(lam, g, f)
```

```python
# Plotting the contour plot for different values of the function f on the x1 and x2 plane
x1 = np.linspace(-1,1,1000)
x2 = np.linspace(-1,1,1000)
X1, X2 = np.meshgrid(x1,x2)
# Now we take some example values of lambda, w0, w1, w2 and plot the contours
w0 = 0
w1 = 1
w2 = 2
lam_0 = 0.2
g = w0 + X1*w1 + w2*X2
f = act(lam_0*g)
fig1 = plt.figure()
ax = fig1.gca()
ax.contour(X1, X2, f)
```
