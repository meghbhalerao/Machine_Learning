```python
import numpy as np
import matplotlib.pyplot as plt
X = np.random.rand(15,20)
X = - 20 + X*(170)
X_bak = X
for i in range(0,X.shape[1]):
    mx = np.max(X[:,i])
    mn = np.min(X[:,i])
    X[:,i] = (X[:,i] - mn)/(mx - mn)
for i in range(0,X.shape[1]):
    plt.plot(X[:,i])
plt.figure()
print(X_bak)
for i in range(0,X_bak.shape[1]):
    plt.plot(X_bak[:,i])
# Binarizing
X = (X>=0).astype(int)
```
